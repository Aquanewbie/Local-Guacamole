```python
import pandas as pd
```

```python
#Read In CSV
Guacamoledf = pd.read_csv('FAOSTAT_data_1-6-2020-Guacamole.csv')
#Rename Agricultural Products to Appropriate GUacamole Ingredients
Guacamoledf=Guacamoledf.replace(to_replace='Anise, badian, fennel, coriander', value='Cilantro', regex=True)
Guacamoledf=Guacamoledf.replace(to_replace='Onions, dry', value='Onions', regex=True)
Guacamoledf=Guacamoledf.replace(to_replace='Onions, shallots, green', value='Onions', regex=True)
Guacamoledf=Guacamoledf.replace(to_replace='Lemons and limes', value='Limes', regex=True)
Guacamoledf
```

```python
Guacamoledfgrouped = Guacamoledf.groupby(['Area', 'Item',]).count
Guacamoledfgrouped()
```

```python
#Find All Coutries in Dataset
Guacamoledf.Area.unique()
```

```python
print (len(Guacamoledf.Area.unique()))
```

```python
#Find Countries that Grow Avocados
AvoCountries = Guacamoledf.loc[Guacamoledf['Item'] == 'Avocados'].Area.unique()
AvoCountries
```

```python
#Check How Many Countries Produce Avocados 
len(AvoCountries)
```

```python
#Create Dictionary Countries that Grow Avocados and Other Guacamole Ingredients
GuacIngredientDict = []
for x in AvoCountries:
    xlist = []
    for i in (Guacamoledf.loc[Guacamoledf['Area'] == x]).Item.unique():
        xlist.append(i)
        xlist.sort()
    GuacIngredientDict.append({'Country':x, 'Produce':xlist})
```

```python
#Show Dictionary of Guacamole Ingredients
GuacIngredientDict

    
```

```python
Guacdf = pd.DataFrame(GuacIngredientDict)
Guacdf
```
